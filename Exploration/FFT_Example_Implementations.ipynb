```python
import sys
import time
import numpy as np  
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

## Fast Fourier Transform (FFT)

This contains some sample FFT implementations and variations



```python
# One dimensional fourier transform on 1 dimenional data

data = np.zeros(100)
for i in range(100):
    data[i] = np.sin((i)*(np.pi)/5)
print(data)

plt.plot(data)
plt.show()

data_f = np.fft.fft(data)
data_mag = abs(data_f)
print(data_mag.shape)

plt.plot(data_mag)
plt.show()
```

```python
# Applying one dimensional and two dimensional FFT on two dimensional data

data_time = np.zeros((10,10), dtype=np.float64)

for i in range(10):
    for j in range(10):
        if j%2==0:
            data_time[i][j]= np.sin((j+1)*(np.pi)/5)
        else:
            data_time[i][j]= np.cos((j+1)*(np.pi)/5)
print(data_time)    

plt.imshow(data_time)
plt.show()
        
data_freq0 = np.fft.fft(data_time,10,0)
data_freq1 = np.fft.fft(data_time,10,1)
data_freq = np.fft.fft2(data_time)

print(data_freq0.shape)
# print(data_freq)
plt.imshow(abs(data_freq0))
plt.show()

print(data_freq1.shape)
# print(data_freq)
plt.imshow(abs(data_freq1))
plt.show()

print(data_freq.shape)
# print(data_freq)
plt.imshow(abs(data_freq))
plt.show()
```

```python
spectral_data_0 = np.sum(data_freq0,axis=0)
spectral_data_1 = np.sum(data_freq1,axis=1)

print(spectral_data_0)
print(spectral_data_1)
```
